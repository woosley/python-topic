# matplotlib -- python作图工具

## 简单画图

```python
%matplotlib inline

import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.show()
```

当传递给plot的参数只有一个的时候，代表Y轴。X轴的数据自动生成。由于python中index以0开始，x轴也以0开始。传递两个参数的时候，分别代表X，Y轴。

```python
plt.plot([1, 2, 3, 4], [5, 7, 8, 10])
plt.show()
```

## 选择画图的样式
plot接受第三个参数，用来表示画线图还是点图，以及颜色，形状等样式。比如

- 点图

```python
x = [1, 2, 3, 4]
y = [5, 7, 8, 10]
plt.plot(x, y, 'ro')
plt.show()
```

- 虚线点图

```python
plt.plot(x, y, '-.gs')
plt.show()
```

`pydoc matplotlib.pyplot.plot`可以查看所有支持的样式。包括线条样式，颜色和点的样式。此外还可以通过kwargs的方式来控制。比如

```python
plt.plot(x, y, linestyle="dashed", marker="o", color='red', linewidth=4)
plt.show()
```

- 添加标签

```python
plt.plot(x, y, label="my line", linestyle="dashed")
plt.show()
## 为什么下图没有标签?
```

## 画板控制
### x y 轴
可以通过`x/ylim，x/yticks`来控制作图范围以及xy轴显示的数值。


```python
plt.plot(x, y, '-o')
plt.xlim(0.5, 5)
plt.xticks([1, 2, 3, 4])
plt.ylim(4, 11)
plt.yticks(y)
plt.show()
```

### 标题，标签和图例
x/ylabel和title函数用来设置xy轴的标签和图片的title
legend用来设置图例


```python
plt.plot(x, y, '-og')
plt.xlim(0.5, 5)
plt.xticks([1, 2, 3, 4])
plt.ylim(4, 11)
plt.yticks(y)
plt.xlabel("this is X")
plt.text(1, 8, "haha") #文本输入
plt.ylabel("this is Y")
plt.title("this is a title")
plt.legend(["yo"])
plt.show()
```

## 多个数据集
plot接受参数，可以在一张图片上面画多个数据集。比如可以画两条线条

```python
plt.plot(x, y, '-go', [12, 11, 10, 3],  [1, 5, 7, 8], '--ro')
plt.show()
```

## 多张图像 
这里杜撰一些概念: 画布和图像 
matplotlib可以在一张__画布__上面画多张图像。而plot.show()可以同时显示多张画布

- plt.figure(n) 表示创建一张画布
- plt.subplot(numrow, numcol, fignum) 表示把当前画布分割成为numrow行，numcol列，以及当前的图像号。fignum可以从1到`numrow*numcol`

以下创建两张画布，每张上面有2个图片

```python
import numpy as np
x = np.linspace(1, 10)
plt.figure(1)
sub1 = plt.subplot(211)
sub1.plot(x, np.sin(x), "--or")
plt.subplot(212)
plt.plot(x, np.cos(x), "--og")
plt.figure(2)
plt.subplot(211)
plt.plot(x, np.tan(x), "--or")
plt.subplot(212)
plt.plot(x, np.arctan(x),"--og")
plt.show()
```

也可以用`subplots`函数在一张画布上面创建多个图像，其返回值是（fig, (aixes))分别表示画布和所有图像。

```python
import numpy as np
x = np.linspace(1, 10)
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(x, np.sin(x), '--ro')
ax1.plot(x, np.cos(x), '--go')
ax2.plot(x, np.tan(x), "--ro")
ax2.plot(x, np.arctan(x), "--go")
plt.show()
```

## 柱状图
柱状图使用`plt.bar`来绘制，纵向图使用`plt.barh`。



```python
height = [4, 3, 5, 7, 15, 25, 3, 9, 7, 2]
left = range(len(height))
plt.bar(left, height,  width=.8, color='green')
plt.show()
```

```python
plt.barh(left, height, color='red')
plt.show()
```

还可以将不同数据画到一起做比较，并显示数据

```python
import numpy as py
age = np.array([10, 20, 30, 40, 50])
weight1 = (40, 60, 70, 75, 63)
weight2 = (35, 45, 50, 44, 51)

fig, ax = plt.subplots()
bar1 = ax.bar(age, weight1, width=4, color='red')
for i in bar1: #显示数据
    ax.text(i.get_x() -1 + i.get_width()/2, 1.05 * i.get_height(), "%d"% i.get_height())
bar2 = ax.bar(age+4, weight2, width=4, color='green')
for i in bar2:
    ax.text(i.get_x() -1 + i.get_width()/2, 1.05 * i.get_height(), "%d"% i.get_height())
ax.set_xticks(age+4)
ax.set_xticklabels(age)
plt.legend(["weight1", "weight2"])
plt.show()
```
