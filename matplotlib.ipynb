# matplotlib -- python作图工具

## 简单画图

```python
%matplotlib inline

import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.show()
```

当传递给plot的参数只有一个的时候，代表Y轴。X轴的数据自动生成。由于python中index以0开始，x轴也以0开始。传递两个参数的时候，分别代表X，Y轴。

```python
plt.plot([1, 2, 3, 4], [5, 7, 8, 10])
plt.show()
```

## 选择画图的样式
plot接受第三个参数，用来表示画线图还是点图，以及颜色，形状等样式。比如

- 点图

```python
x = [1, 2, 3, 4]
y = [5, 7, 8, 10]
plt.plot(x, y, 'ro')
plt.show()
```

- 虚线点图

```python
plt.plot(x, y, '-.gs')
plt.show()
```

`pydoc matplotlib.pyplot.plot`可以查看所有支持的样式。包括线条样式，颜色和点的样式。此外还可以通过kwargs的方式来控制。比如

```python
plt.plot(x, y, linestyle="dashed", marker="o", color='red', linewidth=4)
plt.show()
```

- 添加标签

```python
plt.plot(x, y, label="my line", linestyle="dashed")
plt.show()
## 为什么下图没有标签?
```

## 画板控制
### x y 轴
可以通过`x/ylim，x/yticks`来控制作图范围以及xy轴显示的数值。


```python
plt.plot(x, y, '-o')
plt.xlim(0.5, 5)
plt.xticks([1, 2, 3, 4])
plt.ylim(4, 11)
plt.yticks(y)
plt.show()
```

### title和标签
x/ylabel和title函数用来设置xy轴的标签和图片的title


```python
plt.plot(x, y, '-og')
plt.xlim(0.5, 5)
plt.xticks([1, 2, 3, 4])
plt.ylim(4, 11)
plt.yticks(y)
plt.xlabel("this is X")
plt.ylabel("this is Y")
plt.title("this is a title")
plt.show()
```
